```python
import json
import random
from geopy.geocoders import Nominatim
from turfpy.measurement import bbox, boolean_point_in_polygon
from geojson import Feature, Point, Polygon, MultiPolygon

# Cargar el GeoJSON de Medellín
with open('medellin.json', 'r') as f:
    medellin = json.load(f)
    
# Cargar el GeoJSON de las comunas y corregimientos
with open('barrios.json', 'r') as f:
    barrios = json.load(f)

box = bbox(medellin["features"][0]["geometry"])
poly_aux = [feature["geometry"]["coordinates"] for feature in barrios["features"]]
poly_barrios = MultiPolygon(poly_aux)

# Definir las coordenadas aproximadas de Medellín
lat_range = (box[1], box[3])  # Latitud de Medellín
lng_range = (box[0], box[2])  # Longitud de Medellín

# Inicializar el geolocalizador
geolocator = Nominatim(user_agent="medellin_locator")

# Cargar el GeoJSON de Medellín
with open('medellin.json', 'r') as f:
    medellin = json.load(f)

# Función para obtener la comuna y barrio según las coordenadas
def get_location_info(latitude, longitude):
    location = geolocator.reverse((latitude, longitude), language='es')
    address = location.raw['address']
    comuna = address.get('suburb', 'No encontrado')
    barrio = address.get('neighbourhood', 'No encontrado')
    return comuna, barrio

# Generar puntos de venta únicos
def generate_valid_points(n):
    valid_points = []
    while len(valid_points) < n:
        lat = round(random.uniform(*lat_range), 6)
        lon = round(random.uniform(*lng_range), 6)
        point = Feature(geometry=Point([lon, lat]))
        if boolean_point_in_polygon(point, poly_barrios):
            comuna, barrio = get_location_info(lat, lon)
            if comuna != 'No encontrado' and barrio != 'No encontrado':
                valid_points.append({"latitude": lat, "longitude": lon, "commune": comuna, "neighborhood": barrio})
    return valid_points

# Generar y guardar 100 puntos válidos
points_of_sale = generate_valid_points(100)

with open('valid_points.json', 'w') as file:
    json.dump(points_of_sale, file)

```

Codigo extra para convertir eventos antiguos al formato actual

```python
from datetime import datetime

data = []
with open('simulated_temp.json', 'r') as file:
        for line in file:
            data.append(json.loads(line))
            
with open('sale_points.json', 'r') as f:
    sale_points = json.load(f)
    
def generate_quantity_products(temperature):
    temp_min, temp_max = 10, 30
    normalized_temp = (temperature - temp_min) / (temp_max - temp_min)
    min_quantity, max_quantity = 1, 50
    mean_quantity = normalized_temp * (max_quantity - min_quantity) + min_quantity
    quantity = int(random.gauss(mean_quantity,10))
    return max(min(quantity, max_quantity), min_quantity)

new = []
for i, item in enumerate(data):
    date_now = datetime.strptime(item["date"], "%d/%m/%Y %H:%M:%S")
    point = random.choice(sale_points)
    item["latitude"] = point["latitude"]
    item["longitude"] = point["longitude"]
    item["employee_id"] = point["employees"]["0"] if date_now.weekday() < 5 else point["employees"]["1"]
    item["sale_point_id"] = point["id"]
    item["quantity_products"] = generate_quantity_products(item["current_temperature"])

    del item['event_date']
    del item['partition_date']

    with open('simulated_sales.json', 'a') as file:
        file.write(json.dumps(item) + "\n")

    new.append(item)
    
# with open('generate_sales_1.json', 'w') as f:
#     json.dump(new, f)
```
